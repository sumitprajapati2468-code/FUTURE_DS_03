# Marketing Funnel & Conversion Performance Analysis

This notebook builds the analysis end to end: it creates a realistic synthetic marketing funnel dataset when no CSV is provided, cleans the data, calculates conversion metrics, generates charts, and exports a stakeholder-friendly summary report.

The notebook is intentionally structured so each section explains the reasoning behind the step, not just the code.

## 1) Data Setup

If you already have a funnel CSV, update the `input_csv` path below. Otherwise, the notebook will generate a synthetic six-month dataset with channel variation, funnel attrition, cost, and revenue so the analysis can be tested immediately.

```python
from pathlib import Path
from IPython.display import display, Image

from marketing_funnel_analysis import run_pipeline

base_dir = Path.cwd()
results = run_pipeline(input_csv=None, output_dir=base_dir, force_generate=False)
results['raw_df'].head()
```

```python
display(results['inspection_df'].head(15))
print('Raw rows:', len(results['raw_df']))
print('Cleaned rows:', len(results['cleaned_df']))
print('Unique channels:', results['cleaned_df']['channel'].nunique())
print('Date range:', results['cleaned_df']['date'].min().date(), 'to', results['cleaned_df']['date'].max().date())
```

## 2) Cleaning and Funnel Analysis

The cleaning step standardizes labels, fixes date parsing, fills missing values, removes duplicates, and converts counts/costs to numeric types. Once the data is consistent, the funnel summary and channel performance metrics become trustworthy enough for decision-making.

```python
display(results['funnel_summary'])
display(results['channel_summary'].sort_values('customers', ascending=False))
display(results['weekly_trend'].head())
display(results['monthly_trend'].head())
```

## 3) Visualizations

The funnel chart shows where conversion volume falls away, the channel chart compares traffic volume with conversion rate, and the trend chart shows how efficiency changes over time. These three views are enough to diagnose both structural and time-based issues in the funnel.

```python
display(Image(str(results['chart_paths']['funnel_png'])))
display(Image(str(results['chart_paths']['channel_png'])))
display(Image(str(results['chart_paths']['trend_png'])))
display(Image(str(results['chart_paths']['channel_trend_png'])))
```
